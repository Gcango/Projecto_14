```python
dados2 = dados2.drop("Unnamed: 0", axis=1)
```

```python
#1) Como juntar os datasets afim de obter a nota de cada Estudante?

bd_geral = dados.merge(dados2, on='Name')
bd_geral.head(10)
```

```python
#2) Qual é o estado do Dataset após o merge?
bd_geral.info()
```

```python
#3) - Que medidas podem ser tomadas quanto ao estado do dataset?
bd_geral['Age'].isna().sum()
```

```python
bd_geral['Math Score'].isna().sum()
```

```python
avg_age = bd_geral['Age'].mean()
bd_geral['Age'] = bd_geral['Age'].fillna(avg_age)
bd_geral['Math Score'] = bd_geral['Math Score'].fillna(0)
bd_geral.info()
```

```python
bd_geral.describe()
```

```python
bd_geral.head()
```

```python
bd_geral['City'].unique()
```

```python
bd_geral['City'] = bd_geral['City'].str.replace('Port', 'Pt')
bd_geral['City'].unique()
```

```python
#4) - Como podem estar distribuidos os estudantes quanto ao genero? Que analise se podem fazer?
bd_geral['Gender'].value_counts()
```

```python
bd_geral['Gender'] = bd_geral['Gender'].replace({'Male':'M','Female':'F','Other':'O'})
bd_geral['Gender'].value_counts()
```

```python
#Mostrando os valores em Percentagem%
bd_geral['Gender'].value_counts(normalize=True)
```

```python
#5) - As Classificações e o gerenos podem ser resumidos? Como?
bd_geral.groupby('Gender')['Math Score'].agg(['sum', 'mean'])
```

```python
#6) - Existe alguma forma de observar o tatal das médias 
# (English Score, Math Score e Science Score) em relação ao Genero?

pd.pivot_table(bd_geral, values=['Math Score', 'Science Score', 'English Score'],
               index='Gender', aggfunc=np.mean, margins=True)
```

```python
#6) - É possivel encontrar o total de Score? Como?
#7) - É Possivel encontrar a média por grupo de Genero 

bd_geral['Total Score'] = bd_geral['Math Score'] + bd_geral['English Score'] + bd_geral['Science Score']
bd_geral['Avarege per Group'] = bd_geral.groupby('Gender')['Total Score'].transform('mean')
bd_geral.head()
```

```python
#8) - Que estratégia se pode adotar para indicar se o score do aluno é acima do normal ou normal?
avg_math = bd_geral['Math Score'].mean()
bd_geral['Above Average Math Score'] = np.where(bd_geral['Math Score'] > avg_math, 'Yes', 'No')
bd_geral['Above Average Math Score'].value_counts(normalize=True)
```

```python
#Function Creation
def media_acima(bd_geral, col, new_column):
    col_mean = bd_geral[col].mean()
    bd_geral[new_column] = np.where(bd_geral[col] > col_mean, 'Yes', 'No')
    return bd_geral
```

```python
media_acima(bd_geral, 'Science Score', 'Science Score Above Avg')
media_acima(bd_geral, 'English Score', 'English Score Above Avg')
```
